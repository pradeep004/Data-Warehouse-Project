```sql
USE dw
go 
CREATE TABLE calendar_dim(
    calendar_key INT NOT NULL IDENTITY,
    full_date DATE,
    day_ VARCHAR(15),
    month_ VARCHAR(10),
    quater_ CHAR(2),
    year_ INT,
    PRIMARY KEY(calendar_key)
    WITH (IGNORE_DUP_KEY=ON)
);
GO
SELECT salaries.to_date FROM salaries
INSERT into calendar_dim(full_date,day_,month_,quater_,year_)
select
    salaries.to_date,
    DATEPART(DAY,to_date),
    DATEPART(MONTH,to_date),
    DATEPART(QUARTER,to_date),
    DATEPART(YEAR,to_date)           
FROM
salaries
GO
```

```sql
use dw
go 
CREATE TABLE department_dim(
    deptNo NVARCHAR(50), 
    managerFName NVARCHAR(50),
    managerLName NVARCHAR(50),
    PRIMARY KEY(deptNo)
    WITH(IGNORE_DUP_KEY=ON)
)
INSERT into department_dim(deptNo,managerFName,managerLName)
SELECT dept_manager.dept_no,employees.first_name,employees.last_name
FROM dept_manager
inner JOIN employees
on dept_manager.emp_no = employees.emp_no


```

```sql
USE dw
go
CREATE table employee_dim(
    empNo INT,
    firstName NVARCHAR(50),
    lastName NVARCHAR(50),
    gender NVARCHAR(50),
    birthDate NVARCHAR(50),
    hiredDate DATE,
    title NVARCHAR(50),
    PRIMARY KEY(empNo)
    WITH(IGNORE_DUP_KEY=ON)
)
INSERT into employee_dim(empNo,firstName,lastName,gender,birthDate,hiredDate,title)
SELECT employees.emp_no,employees.first_name,employees.last_name,employees.gender,employees.birth_date,employees.hire_date,titles.title
from employees
INNER JOIN titles
on employees.emp_no = titles.emp_no
```
